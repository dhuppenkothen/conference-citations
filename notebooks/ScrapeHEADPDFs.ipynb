## Scraping PDFs

Let's scrape us some PDFs.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("whitegrid")
sns.set_palette('colorblind')

import numpy as np
import pandas as pd

import PyPDF2
```

Let's try the simple text data:

```python
f=open("../data/Head/head_15_abstracts.txt", 'r')

data = []
for line in f:
    data.append(line)
```

Let's ignore the first lines, which just define the different sessions:

```python
data = data[75:]
```

```python
data
```

```python
authors = []
first_author = []

for i, d in enumerate(data):
    if "Author(s)" in d:
        authors.append(d)
        a = d.split(": ")[1]
        fa = a.split(", ")[0]
        if fa[-1] == "\n":
            fa = fa[:-1]
        if fa[-1] in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
            fa = fa[:-1]
        if fa[-1] == " ":
            fa = fa[:-1]

        first_author.append(fa)          
```

```python
first_author
```

Hooray! I can extract authors. But we want to associate them with posters and talks. Let's load the talks document:

```python
talk_ids = []

f2 = open("../data/Head/Talks", "r")
for line in f2:
    t = line.split("–")
    
    talk_ids.append(int(t[0]))
```
